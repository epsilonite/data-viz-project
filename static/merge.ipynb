```python
import pandas as pd
import geopandas as gpd
```

```python
faostat=gpd.read_file('data/geojson/faostat.geojson')
iucn=gpd.read_file('data/geojson/iucn.geojson')
```

```python
faostat_ = faostat.set_index('name')
```

```python
faostat_.columns
```

```python
iucn_ = iucn.set_index('name')[['endangered']]
```

```python
iucn_.columns
```

```python
merge = gpd.GeoDataFrame(faostat_.join(iucn_),geometry='geometry')
merge
```

```python
merge.to_file('data/geojson/merged.geojson',driver='GeoJSON')
```
