```python
QCL_country = QCL_df.groupby('Country').agg(list)
```

```python
QCLdictC = {}
```

```python
for name in QCL_country.index:
    dfx = pd.DataFrame([*QCL_country.loc[name]], index=['Name','Year','Value']).T
    dfx = dfx.groupby('Name').agg(list)
    items=[]
    for n in dfx.index:
        items.append({'item':n,'years':[*dfx.Year[n]],'values':[*dfx.Value[n]]})
    QCLdictC[name] = items
```

```python
gdf = pd.DataFrame({'name': QCLdictC.keys(), 'items':QCLdictC.values()})
gdf
```
