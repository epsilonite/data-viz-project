```python
import pandas as pd
import geopandas as gpd
```

```python
gdf = gpd.read_file('data/geometry/world-administrative-boundaries.geojson')
```

```python
df = pd.read_json('data/faostat/faostat.json').T
```

```python
hash = pd.read_csv('data/faostat/hashtable.csv')
```

```python
fao_drop = ['Bonaire, Sint Eustatius and Saba', 'Curaçao','Sint Maarten (Dutch part)']
```

```python
df.drop(fao_drop,inplace=True)
```

```python
hash.columns
```

```python
fao_hash=hash.set_index('faostat').rename(columns={'map':'name'})
```

```python
fao_hash
```

```python
df1 = df.join(fao_hash).set_index('name')
df1
```

```python
gdf1 = gdf[['name','continent','region','status','geometry']].sort_values('name').set_index('name')
gdf1
```

```python
merge = gpd.GeoDataFrame(df1.join(gdf1)[['region','continent','status','landarea','landuse','items','geometry']],geometry='geometry')
merge.rename({'map':'name'},inplace=True)
merge
```

```python
merge['landuse'] = [str(x) for x in merge.landuse.values]
merge['items'] = [str(x) for x in merge['items'].values]
```

```python
merge.to_file('data/geojson/faostat.geojson', driver='GeoJSON')
```
