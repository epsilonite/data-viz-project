```python
import pandas as pd
import requests
```

```python
df = pd.read_csv('data/iucn/iucn_init2.csv')
```

```python
# Function to get taxonomic details from GBIF API
def get_taxonomy_info(scientific_name):
    url = f"https://api.gbif.org/v1/species/match?name={scientific_name}"
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        # Extracting phylum, class, and order
        phylum = data.get('phylum')
        tax_class = data.get('class')
        order = data.get('order')
        genus = data.get('genus')
        species = data.get('species')
        return phylum, tax_class, order,genus,species
    else:
        return None, None, None
```

```python
sci_dict = {}
for n in df.sci_name.unique():
    phylum, tax_class, order = get_taxonomy_info(n)
    sci_dict[n] = {'phylum':phylum,'class':tax_class,'order':order}
```

```python
pd.DataFrame(sci_dict).to_csv('data/iucn/gbif_taxonomy.csv')
```
