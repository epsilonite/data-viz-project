```python
import pandas as pd
import re
```

## Load IUCN point data (smaller)
---

```python
iucn_df1 = pd.read_csv('data/iucn/points_data1.csv')
```

```python
iucn_df1.columns
```

```python
iucn_df11 = pd.DataFrame()
```

```python
iucn_df11['sci_name'] = iucn_df1.sci_name.astype('string')
iucn_df11['legend'] = iucn_df1.legend.astype('string')
iucn_df11['source'] = iucn_df1.source.fillna(' ').astype('string')
iucn_df11['event_year'] = iucn_df1.event_year.fillna(0).astype('Int64')
iucn_df11['longitude'] = iucn_df1.longitude
iucn_df11['latitude'] = iucn_df1.latitude
```

```python
iucn_df11.dtypes
```

```python
iucn_df11.sci_name.str.len().max()
```

```python
iucn_df11.legend.str.len().max()
```

```python
iucn_df11.source.str.len().max()
```

```python
iucn_df11.legend.unique()
```

```python
for i,row in iucn_df11.iterrows():
    if row.event_year==0:
        nums = re.findall(r'\d+', row.source)
        if len(nums)>0:
            for j in nums:
                if len(j)==4: 
                    iucn_df11.loc[i,'event_year'] = int(j);
                    break;
```

```python
for i,row in iucn_df11.iterrows():
    if row.event_year == 0:
        if (row.source[0:6] =='Global') | (row.source.lower() == 'gbif'): iucn_df11.loc[i,'event_year'] = 2019
        if row.source[12:23] =='inaturalist': iucn_df11.loc[i,'event_year'] = 2020
        if row.source in ['Kearney','Kearney & Hoffmann','Kearney, Hoffmann, Hossain','Kearney, Hossain',
                            'Kearney, Song, Hossain & Sinclair','Hoffmann']: iucn_df11.loc[i,'event_year'] = 2019
        if row.source == 'LepiMAP, SAM-LEP': iucn_df11.loc[i,'event_year'] = 1996
        if row.source == 'LepiMAP, ABRI': iucn_df11.loc[i,'event_year'] = 1998
        if row.source == 'LepiMAP, SABCA-LIT': iucn_df11.loc[i,'event_year'] = 1991
```

```python
iucn_df00 = iucn_df11.loc[iucn_df11.event_year==0]
```

```python
iucn_df00.count()
```

```python
iucn_df00.source.unique()
```

```python
iucn_df11[['sci_name','legend','event_year','longitude','latitude']].to_csv('data/iucn/iucn_init1.csv',index=False)
```

```python
iucn_df11.loc[iucn_df11.event_year>0].count()
```
